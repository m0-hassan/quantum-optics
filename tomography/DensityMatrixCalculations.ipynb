```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
#define the pauli matrices
identity = np.array([[1,0],
                  [0,1]])

pauliX = np.array([[0,1],
                  [1,0]])

pauliY = np.array([[0,-1j],
                  [1j,0]])

pauliZ = np.array([[1,0],
                  [0,-1]])
```

# Trial 1 Calculation

```python
#photon count data for trial 1
cH1 = 461
cV1 = 7.333333
cPlus1 = 222
cMinus1 = 220.333333
cRight1 = 228.666667
cLeft1 = 193.333333

#QST data for trial 1
aX1 = 0.003752
aY1 = 0.079540
aZ1 = 1.021261

#density matrix calculation
trial1Matrix = 1/2*(identity + aX1*pauliX + aY1*pauliY + aZ1*pauliZ)
print("The density matrix and trace are")
print(trial1Matrix)
print(np.trace(trial1Matrix))
print()

#stoke vector calculation
s0 = cH1 + cV1
s1 = cPlus1 - cMinus1
s2 = cRight1 - cLeft1
s3 = cH1 - cV1
stokeVector1 = np.array([s0,s1,s2,s3])

print("The Stoke vector is:")
print(stokeVector1)
print()


#bloch sphere vector calculation
blochVector1 = 1/s0*np.array([s1,s2,s3])
print("The Bloch sphere vector is:")
print(blochVector1)
```
